**Задание 2: Создайте функцию, которая принимает как аргументы целое число N и первый элемент (вещественное число), и разность (вещественное число) и создает матрицу numpy по диагонали, которой располагаются первые N членов арифметической прогрессии.**

```python
def make_diag_matrix(N: int, first_element: float, difference: float):
    last_element = first_element + difference * N  # последний элемент последовательности (без учета шага между элементами)
    elements = np.arange(first_element, last_element, difference)  # арифметическая последовательность
    matrix = np.diag(elements)  # матрица с элементами по диагонали
    return matrix


ap_matrix = make_diag_matrix(5, 10.0, 5.0)
print(ap_matrix)
```

**Задание 5: Создайте две матрицы размером (5,5). Одна матрица содержит 5 в шахматном порядке как в задаче домашнего задания, другая имеет треугольную форму содержащую 5 на основной диагонали и в позициях выше ее, а ниже все 0. Посчитайте их детерминант и найдите обратные матрицы.**

```python
chess_matrix = np.zeros((5, 5))  # двумерный массив 5*5 из нулей
chess_matrix[::2, ::2] = 5  # Заполняем клетки с четными индексами пятерками
chess_matrix[1::2, 1::2] = 5  # Заполняем клетки с нечетными индексами пятерками
print("Шахматная матрица:\n", chess_matrix)

triangle_matrix = np.ones((5, 5))  # двумерный массив 5*5 из единиц
triangle_matrix = triangle_matrix * 5  # заполним массив пятерками
triangle_matrix = np.triu(triangle_matrix)  # все что ниже диагонали заполним нулями
print("Треугольная матрица:\n", triangle_matrix)

# посчитаем детерминанты матриц
det_chess = np.linalg.det(chess_matrix)
det_triangle = np.linalg.det(triangle_matrix)
print(f"Детерминант шахматной матрицы: {det_chess}")
print(f"Детерминант треугольной матрицы: {det_triangle}")
print()

# Обратные матрицы
try:
    inv_chess = np.linalg.inv(chess_matrix)
    print("Матрица, обратная шахматной:\n", inv_chess)
except np.linalg.LinAlgError:
    print("Для шахматной матрицы не существует обратной")
print()

try:
    inv_triangle = np.linalg.inv(triangle_matrix)
    print("Матрица, обратная треугольной:\n", inv_triangle)
except np.linalg.LinAlgError:
    print("Для треугольной матрицы не существует обратной")
```

**Задание 17: Постройте круговую диаграмму для признака Rooms, иллюстрирующую количество квартир в процентах в зависимости от количества комнат. Сделайте сектор с наибольшим числом квартир выдвинутым.**
